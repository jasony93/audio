```python
import librosa
import librosa.display
import scipy as sp
import IPython.display as ipd
import matplotlib.pyplot as plt
import numpy as np
```

```python
audio_path = "audio_data/test/M_C_S_jyyang_0108_74.wav"
ipd.Audio(audio_path)
```

```python
signal, sr = librosa.load(audio_path, sr=None)
plt.figure(figsize=(18,8))
librosa.display.waveplot(signal, sr=sr, alpha=0.5)
plt.show()
```

# Derive spectrum using FT 

```python
ft = sp.fft.fft(signal)
magnitude = np.absolute(ft)
frequency = np.linspace(0, sr, len(magnitude))

plt.figure(figsize=(18,8))
plt.plot(frequency[:1000], magnitude[:1000])
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.show()
```

```python
samples = range(len(signal))
t = librosa.samples_to_time(samples, sr=sr)

plt.figure(figsize=(18, 8))
plt.plot(t[10000:10400], signal[10000:10400])
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()
```

```python
f = 220
phase = 0.8

sin = 0.01 * np.sin(2*np.pi * (f * t - phase))
plt.figure(figsize=(18, 8))
plt.plot(t[10000:10400], sin[10000:10400], color='r')
plt.plot(t[10000:10400], signal[10000:10400], color='b')
plt.fill_between(t[10000:10400], 10*sin[10000:10400]*signal[10000:10400], color="y")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()
```
