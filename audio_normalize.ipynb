# Generate normalized audio

```python
from pydub import AudioSegment, effects

file_name = 'Arduino_Small'

raw_sound = AudioSegment.from_file( file_name + '.wav', 'wav')
normalized_sound = effects.normalize(raw_sound)
normalized_sound.export(file_name + '_Normalized.wav', format='wav')
```

# Compare normalized and raw data 

```python
import matplotlib.pyplot as plt
import numpy as np
import wave
import sys

file_name1 = 'Mobile_Large'
file_name2 = 'Mobile_Large_Normalized'
file_name3 = 'Arduino_Small'
file_name4 = 'Arduino_Small_Normalized'

fig = plt.figure(figsize=(15,15))

def plot_wav(file_name, index):
    spf = wave.open(file_name + '.wav', "r")

    # Extract Raw Audio from Wav File

    # n frames of audio to bytes where n = -1
    signal = spf.readframes(-1)

    # A new 1-D array initialized from text data in a string
    signal = np.frombuffer(signal, "Int16")
    # print(signal)

    # getnchannel = 1 => mono  /  getnchannel = 2 => stereo
    if spf.getnchannels() == 2:
        print("Just mono files")
        sys.exit(0)

    global fig
    subplt = plt.subplot(2, 2, index)
    # ax = fig.add_subplot(111)
#     subplt.spines['bottom'].set_color('white')
#     subplt.tick_params(axis='x', colors='white')
    # y = fig.add_subplot(111)
#     subplt.spines['left'].set_color('white')
#     subplt.tick_params(axis='y', colors='white')

    plt.title(file_name, fontdict={'fontsize': 20, 'color': 'black'})
    plt.plot(signal)
    fig.add_subplot(subplt)
    # plt.show()

plot_wav(file_name1, 1)
plot_wav(file_name2, 2)
plot_wav(file_name3, 3)
plot_wav(file_name4, 4)
plt.show()
```
