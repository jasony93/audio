```python
import librosa
import librosa.display
import os
import soundfile as sf
import matplotlib as plt

data, sr = librosa.load('C://Users/jason/jupyter/audio/audio_data/test/신정례_가까운거_test.wav', sr=None)
shifted = librosa.effects.pitch_shift(data, sr, n_steps=10)
stretched = librosa.effects.time_stretch(data, 1.2)

X_original = librosa.stft(data.astype('float'))
Xdb_original = librosa.amplitude_to_db(X_original)

X_shifted = librosa.stft(shifted.astype('float'))
Xdb_shifted = librosa.amplitude_to_db(X_shifted)

X_stretched = librosa.stft(stretched.astype('float'))
Xdb_stretched = librosa.amplitude_to_db(X_stretched)

# X_stretched = librosa.stft()
# fig = plt.figure(figsize=(25,25))
# subplt = plt.subplot(2,2,1)
# plt.plot(Xdb_original)
librosa.display.specshow(Xdb_original[:,:50], sr=sr, x_axis='time', y_axis='hz')

# plt.show()
```
